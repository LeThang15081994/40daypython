```python
#Đọc ảnh 1.jpg và hiển thị bằng matplotlib và opencv

img = cv.imread("1.jpg")
plt.imshow(img,cmap ="viridis")
plt.show()
```

```python
# chuyển từ ảnh màu sang ảnh xám
img_gray = cv.cvtColor(img,cv. COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap="gray")
plt.show()
```

```python
# tăng ảnh gốc lên 50 đơn vị và giảm ảnh gốc xuống 80 đơn vị sử dụng Opencv
brighness_img = cv.add(img,50)
dark_img = cv.subtract(img,80)

ax1 = plt.subplot(221)
ax1.imshow(img[::-1])
ax1.set_title('ORIGIN')

ax1 = plt.subplot(222)
ax1.imshow(dark_img[::-1])
ax1.set_title('DARK IMAGE')

ax1 = plt.subplot(223)
ax1.imshow(brighness_img[::-1])
ax1.set_title('BRIGHNEES IMAGE')

'''axs[0, 0].imshow(img[::-1])
axs[0, 0].set_title('ORIGIN')

axs[0, 1].imshow(dark_img[::-1])
axs[0, 1].set_title('DARK IMAGE')

axs[1, 0].imshow(brighness_img[::-1])
axs[1, 0].set_title('BRIGHNEES IMAGE')'''

plt.show()
```
