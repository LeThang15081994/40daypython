Bộ lọc Gauss và hàm Gauss

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#Load img
img = cv.imread("2.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
'''cv.imshow("img",img)
cv.waitKey(0)
cv.destroyWindow()'''
plt.imshow(img,cmap="gray")
plt.show()
```

```python
#Bộ lọc Gaussian trong opencV
img1 = cv.GaussianBlur(img,(5,5),2.5)
img2 = cv.GaussianBlur(img,(5,5),5)
img3 = cv.GaussianBlur(img,(5,5),10)

ax1 =plt.subplot(221)
ax1.imshow(img1,cmap="gray")
ax1.set_title("Origin")

ax2 =plt.subplot(222)
ax2.imshow(img2,cmap="gray")
ax2.set_title("Sigma = 2.5")

ax3 =plt.subplot(223)
ax3.imshow(img1,cmap="gray")
ax3.set_title("Sigma = 5")

ax4 =plt.subplot(224)
ax4.imshow(img3,cmap="gray")
ax4.set_title("Sigma = 10")
plt.show()
```

```python
#Cách 2

import numpy as np
```

```python
def gaussian_kernel(size, sigma):
    if size % 2 == 0:
        size = size + 1

    max_point = size // 2  # Điểm tối đa trong cả hai hướng (x,y)
    min_point = -max_point  # Điểm tối thiểu trong cả hai hướng (x,y)

    # Ma trận kernel
    K = np.zeros((size, size))

    # Tính toán giá trị cho mỗi ô trong ma trận kernel
    for x in range(min_point, max_point + 1):
        for y in range(min_point, max_point + 1):
            # Công thức phân phối Gaussian
            value = np.exp(-(x**2 + y**2) / (2 * sigma**2)) / (2 * np.pi * sigma**2)
            K[x - min_point, y - min_point] = value

    return K
```

```python
# Tạo kernel Gaussian có kích thước 5x5 và độ lệch chuẩn là 1.4
kernel = gaussian_kernel(5, 10)

# Đọc hình ảnh

# Áp dụng ma trận kernel vào hình ảnh để làm mờ
img_gaussian = cv.filter2D(img, -1, kernel)
```

```python
ax1 =plt.subplot(221)
ax1.imshow(img,cmap="gray")
ax1.set_title("Origin")


ax2 =plt.subplot(222)
ax2.imshow(img_gaussian,cmap="gray")
ax2.set_title("Gaussian_img")
```
